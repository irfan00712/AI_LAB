```python
df = pd.read_csv('A.csv')
print(df)
```

```python
new_df = df.dropna()

print(new_df.to_string())
```

```python
new_df = df.fillna(50)
print(new_df)
```

```python
x = df["weight"].mean()
print("Mean is",x)
new_df=df.fillna(x)
print(new_df)
```

```python
x = df["weight"].median()
print("median is",x)
new_df=df.fillna(x)
print(new_df)
```

```python
x = df["weight"].mode()[0]
print("Mode is",x)
new_df=df.fillna(x)
print(new_df)
```
